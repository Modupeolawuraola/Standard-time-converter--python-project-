```python
from datetime import datetime, timedelta
from pytz import timezone
import pytz

print(" DATE/TIME INPUT MUST BE STRICTLY ENTERED IN NUMBER FORMAT")

utc = pytz.utc
state = input('specify is your country location: ').title()
userclass=input('specify your class program: format Data Analysis, Data Science/ML: ')
userCohort= input('specify your  bootcamp cohort:format cohort xx: ')
bootcampMonth = int(input("specify month of the bootcamp schedule: format xx: "))
bootcampDay = int(input("specify your day of bootcamp schedule: format xx: "))
bootcampHour = int(input("specify bootcamp time schedule in (hour): format xx: "))
bootcampMinute = int(input("specify bootcamp time schedule in (min) : format xx:"))
def timeConverter():
    for country in countries:
        if state in country:
            
            myCountry = timezone(country)
            countryTimeZoneList = pytz.timezone(country)
            myCountryCurrentTime = datetime.now(countryTimeZoneList)
            standardTime = timezone('America/New_York')
            print(f"{country} currrent time/timezone is {myCountryCurrentTime}")
            fmt = '%m-%d %H:%M: %Z%z'
            loc_dt = standardTime.localize(datetime(bootcampMonth, bootcampDay, bootcampHour, bootcampMinute))
            loc_dt.strftime(fmt)
            ams_dt = loc_dt.astimezone(myCountry)
            print(f"Your lecture converted time/timezone is {ams_dt.strftime(fmt)}")
            
timeConverter()
```
